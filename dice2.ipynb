<center><H1>Dice Problem</H1></center>

We have 5 dice in a box: 4, 6, 8, 12 and 20-sided dice. Answer the following questions using
Bayesian tables:

Someone picks a die at random, roll it twice and just informs you the results are 9 and 5. Find the probability that the die is 4, 6, 8, 12 and 20-sided.

```python
import pandas as pd
from fractions import Fraction
```

```python
table = pd.DataFrame(index=[4, 6, 8, 12, 20])
table['prior'] = Fraction(1, 5)
table['likelihood1'] = 0, 0, 0, Fraction(1, 12), Fraction(1, 20)
table['likelihood2'] = 0, Fraction(1, 6), Fraction(1, 8), Fraction(1, 12),  Fraction(1, 20)
table['likelihood'] = table['likelihood1'] * table['likelihood2']
table
```

Once we have priors and likelhoods, the remaining steps are always the same, so we will put them in a function:

```python
def update(table):
    """Compute the posterior probabilities."""
    table['unnorm'] = table['prior'] * table['likelihood']
    prob_data = table['unnorm'].sum()
    table['posterior'] = table['unnorm'] / prob_data
    return prob_data
```

```python
prob_data=update(table)
table
```

```python
table['posterior_float'] = table['posterior'].apply(lambda x: round(float(x), 4))
table
```
